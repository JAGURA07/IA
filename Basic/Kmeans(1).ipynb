```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
# Leer los datos
df = pd.read_csv('old_faithful.csv')
X_std = StandardScaler().fit_transform(df)
k=2
```

```python
def k_means(X, k, centros=None, iter=100):
    if centros is None:
        rnd_centros = np.random.choice(np.arange(X.shape[0]), k, replace=False)
        centros = X[rnd_centros]
    for _ in range(iter):
        distancias = np.sum(np.sqrt((X - centros[:, np.newaxis]) ** 2), axis=-1)
        cluster = np.argmin(distancias, axis=0)
        for i in range(k):
            msk = (cluster == i)
            centros[i] = np.mean(X[msk], axis=0) if np.any(msk) else centros[i]

    return cluster, centros
```

```python
k_means(X_std, k, None, iter=100)
```

```python
clusters,centroides=k_means(X_std, k, None, iter=100)
```

```python
plt.scatter(X_std[:, 0], X_std[:, 1],c=clusters)
plt.plot(centroides[:, 0], centroides[:, 1],'*r')
```
